### t検定

- 帰無仮説 H_0：推薦の有無によるタグ付け数の差がない．
- 対立仮説 H_1：推薦の有無によるタグ付け数の差がある．


```python
yes = [32,27,19,31,14]
no = [8,19,16,18,14]
m_yes,m_no = sum(yes)/5,sum(no)/5
print(m_yes,m_no)
lst = [ x-y for x,y in zip(yes,no)]
print(lst)
print(m_yes-m_no)
```

```python
# t分布の計算
from scipy.stats import t
import numpy as np

# グラフ可視化
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams["font.family"]     = "sans-serif"
rcParams["font.sans-serif"] = "Hiragino Maru Gothic Pro"



"""
****************************************
t分布
****************************************
"""
# グラフ設定
fig,ax = plt.subplots(1,1, figsize=(10,8))

# X軸
X = np.linspace(-5,5,100)

# 自由度 n = 1〜5までを用意
n_deg = [4]

# 自由度 n=1〜5までのグラフを描画
for i in n_deg:
    ax.plot(X, t(i).pdf(X), linestyle='-', label="自由度 n={}".format(i))


"""
****************************************
グラフ書式設定
****************************************
"""    
plt.xlim(-5, 5)              # X軸範囲
plt.ylim(0, 0.4)             # Y軸範囲
plt.xticks(fontsize= 12)     # X軸フォントサイズ
plt.yticks(fontsize= 12)     # Y軸フォントサイズ
plt.title("t分布", size=16)   # タイトル
plt.legend()                 # 凡例
plt.show()
```
